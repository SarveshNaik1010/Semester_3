## B. Feature Engineering

```python
# Handling the missing values. 
# Replacing them with the mode of the particular column (NaN)
for column in structuredData.columns:
    mode_value = structuredData[column].mode()[0]  # Get the mode of the column
    structuredData[column].fillna(mode_value)  # Replace NaN with the mode
structuredData
```

```python
# Values Encoding
# Altering the column values of 'Manufacturer Country' by abbreviating them
country_abbreviations = {
    "India": "IND",
    "Italy": "ITY",
}
structuredData['Manufacturer Country'] = structuredData['Manufacturer Country'].replace(country_abbreviations)
structuredData
    
```

```python
# Label encoding
# Changes the label name by using the replace function (Not permanent)
structuredData.rename(columns={"Carbohydrates": "Carbs", "Total Sugars": "Net Sugar", "Calories per Serving": "Calories/Serving"})
```
