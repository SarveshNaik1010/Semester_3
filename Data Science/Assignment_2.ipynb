````python
import pandas as pd

# Fetching data
bengaluruHouseData = pd.read_csv("./Bengaluru_House_Data.csv");
bengaluruHouseData
````

### A. Data for formats of ```merging```, ```grouping``` and ```concatenation```

````python
# 1. Data Wrangling

# Missing field count
nullFieldsCount = bengaluruHouseData.isnull().sum()
print(nullFieldsCount)
````

````python
newBengaluruHouseData = bengaluruHouseData.fillna({'society': 'ArsiaEx', 'bath': '2.0', 'size': "2BHK" })
newBengaluruHouseData
````

# Merging

````python
# Creating subset with different columns
tableHalf1 = newBengaluruHouseData[['area_type','availability','location','size']]
tableHalf2 = newBengaluruHouseData[['area_type','society','total_sqft','bath', 'balcony', 'price']]

tableHalf1.merge(tableHalf2)
````

````python
tableHalf1.merge(tableHalf2, how='outer')
````

````python
tableHalf1.merge(tableHalf2, how='right')
````

````python
tableHalf1.merge(tableHalf2, how='left')
````

# Grouping

````python
print("Total price of all houses with area_type categoery wise area_type")
pd.DataFrame(newBengaluruHouseData.groupby("area_type")['price'].sum())
````

````python
newHouses = pd.DataFrame([{
    "area_type": "Plot Area",
    "availability": "Ready To Move",
    "location": "Bagalur Main Road",
    "size": "4 Bedroom",
    "society": 'Vantage', 
    "total_sqft": 3600,
    "bath": 4,
    "balcony": 1.0, 
    "price": 300
}, {
    "area_type": "Plot Area",
    "availability": "Ready To Move",
    "location": "Bagalur Main Road",
    "size": "2 Bedroom",
    "society": 'Vantage', 
    "total_sqft": 3600,
    "bath": 3,
    "balcony": 1.0, 
    "price": 200
}])

pd.concat([newBengaluruHouseData, newHouses], axis=0)
````
